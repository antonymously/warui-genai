```python
import os
import json
from dotenv import load_dotenv

load_dotenv()
```

```python
import requests
```

```python
RENSHUU_API_KEY = os.getenv('RENSHUU_API_KEY')
```

```python
headers = {
    'Authorization': f'Bearer {RENSHUU_API_KEY}',
    'Content-Type': 'application/json'
}

response = requests.get(
    "https://api.renshuu.org/v1/profile", 
    headers=headers
)
print(json.dumps(response.json(), indent = 4))
```

```python
# lists, I don't use this in renshuu

headers = {
    'Authorization': f'Bearer {RENSHUU_API_KEY}',
    'Content-Type': 'application/json'
}

response = requests.get(
    "https://api.renshuu.org/v1/lists", 
    headers=headers
)
print(json.dumps(response.json(), indent = 4))
```

```python
# schedules - I use this more in renshuu
# lists, I don't use this in renshuu
headers = {
    'Authorization': f'Bearer {RENSHUU_API_KEY}',
    'Content-Type': 'application/json'
}

response = requests.get(
    "https://api.renshuu.org/v1/schedule", 
    headers=headers
)
print(json.dumps(response.json(), indent = 4))
```

```python
# schedule id for N4 grammar: 4419328
# list the terms in this schedule
# this is good! it includes the mastery

headers = {
    'Authorization': f'Bearer {RENSHUU_API_KEY}',
    'Content-Type': 'application/json'
}

response = requests.get(
    "https://api.renshuu.org/v1/schedule/5070559/list", 
    headers=headers,
    params = {
        "pg": 1
    }
)
print(json.dumps(response.json(), indent = 4))
```
