```python
# NOTE: testing mastery perc weights
test_terms = [
    {
        "term": "a",
        "mastery_avg_perc": 100,
    },
    {
        "term": "b",
        "mastery_avg_perc": 20,
    },
    {
        "term": "b",
        "mastery_avg_perc": 55,
    },
]

def get_weights(terms):
    min_weight = 10
    weights = [max(100 - t["mastery_avg_perc"], 0) + min_weight for t in terms]
    
    sum_weights = sum(weights)
    weights = [w/sum_weights for w in weights]

    return weights

get_weights(test_terms)
```
